```python
from collections import deque

muc_tieu = [
    [1, 2, 3],
    [8, '.', 4],
    [7, 6, 5]
]

bat_dau = [
    [2, 8, 3],
    [1, 6, 4],
    [7, '.', 5]
]

def ma_hoa(trang_thai):
    return str(trang_thai)

def tim_o_trong(trang_thai):
    for i in range(3):
        for j in range(3):
            if trang_thai[i][j] == '.':
                return i, j

def bfs(trang_thai_dau):
    queue = deque([(trang_thai_dau, [trang_thai_dau])])
    
    da_tham = set()
    da_tham.add(ma_hoa(trang_thai_dau))

    huong_di = [
        (-1, 0),  # LÊN
        (1, 0),   # XUỐNG
        (0, -1),  # TRÁI
        (0, 1)    # PHẢI
    ]

    while queue:
        trang_thai, duong_di = queue.popleft()

        if trang_thai == muc_tieu:
            print(f"🎉 ĐÃ TÌM THẤY MỤC TIÊU SAU {len(duong_di) - 1} BƯỚC DI CHUYỂN!\n")
            for index, buoc in enumerate(duong_di):
                print(f"--- Bước {index} ---")
                for hang in buoc:
                    print(hang)
                print()
            return True

        x, y = tim_o_trong(trang_thai)

        for dx, dy in huong_di:
            nx, ny = x + dx, y + dy

            if 0 <= nx < 3 and 0 <= ny < 3:
                trang_thai_moi = [hang[:] for hang in trang_thai]
                trang_thai_moi[x][y], trang_thai_moi[nx][ny] = trang_thai_moi[nx][ny], trang_thai_moi[x][y]

                ma_moi = ma_hoa(trang_thai_moi)
                if ma_moi not in da_tham:
                    da_tham.add(ma_moi)
                    queue.append((trang_thai_moi, duong_di + [trang_thai_moi]))

    print("Không tìm thấy đường đi giải bài toán.")
    return False

bfs(bat_dau)
```

```python
muc_tieu = [
    [1, 2, 3],
    [8, '.', 4],
    [7, 6, 5]
]

bat_dau = [
    [2, 8, 3],
    [1, 6, 4],
    [7, '.', 5]
]

def tim_o_trong(trang_thai):
    for i in range(3):
        for j in range(3):
            if trang_thai[i][j] == '.':
                return i, j

def dfs_gioi_han_stack(trang_thai_dau, gioi_han_do_sau):
    stack = [(trang_thai_dau, [trang_thai_dau])]
    
    huong_di = [
        (-1, 0),  # LÊN
        (1, 0),   # XUỐNG
        (0, -1),  # TRÁI
        (0, 1)    # PHẢI
    ]

    while stack:
        trang_thai, duong_di = stack.pop()

        if trang_thai == muc_tieu:
            return duong_di

        do_sau_hien_tai = len(duong_di) - 1
        if do_sau_hien_tai < gioi_han_do_sau:
            x, y = tim_o_trong(trang_thai)

            for dx, dy in reversed(huong_di):
                nx, ny = x + dx, y + dy

                if 0 <= nx < 3 and 0 <= ny < 3:
                    trang_thai_moi = [hang[:] for hang in trang_thai]
                    trang_thai_moi[x][y], trang_thai_moi[nx][ny] = trang_thai_moi[nx][ny], trang_thai_moi[x][y]

                    if trang_thai_moi not in duong_di:
                        stack.append((trang_thai_moi, duong_di + [trang_thai_moi]))

    return None

def iddfs(trang_thai_dau):
    do_sau_max = 0
    
    while True:
        loi_giai = dfs_gioi_han_stack(trang_thai_dau, do_sau_max)
        
        if loi_giai is not None:
            print(f"🎯 DFS (STACK) ĐÃ TÌM THẤY MỤC TIÊU Ở ĐỘ SÂU TẦNG: {do_sau_max}\n")
            for index, buoc in enumerate(loi_giai):
                print(f"--- Bước {index} ---")
                for hang in buoc:
                    print(hang)
                print()
            return True
        
        do_sau_max += 1

iddfs(bat_dau)
```
